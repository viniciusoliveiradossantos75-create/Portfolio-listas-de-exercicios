```python
# 09 Conversor de Temperatura Completo
#Leia uma temperatura e a unidade de origem usando match/case ("C" para Celsius, "F" para Fahrenheit ou "K" para Kelvin).
#Valide que temperaturas em Kelvin não sejam negativas.
#Calcule e exiba a temperatura convertida nas outras duas escalas.
#Use variáveis descritivas como temp_celsius, temp_fahrenheit e temp_kelvin. Use case _ para unidade inválida.

valor = float(input("Digite o valor da temperatura: "))
unidade = input("Digite a unidade de origem (C, F ou K): ").upper()

    # Usando match/case para identificar a unidade
match unidade:
        case "C":
            # Celsius para Fahrenheit e Kelvin
            fahrenheit = (valor * 9/5) + 32
            kelvin = valor + 273.15
            print(f"\n{valor}°C equivale a:")
            print(f"{fahrenheit:.2f}°F")
            print(f"{kelvin:.2f}K")

        case "F":
            # Fahrenheit para Celsius e Kelvin
            celsius = (valor - 32) * 5/9
            kelvin = (valor - 32) * 5/9 + 273.15
            print(f"\n{valor}°F equivale a:")
            print(f"{celsius:.2f}°C")
            print(f"{kelvin:.2f}K")

        case "K":
            # Kelvin para Celsius e Fahrenheit
            celsius = valor - 273.15
            fahrenheit = (valor - 273.15) * 9/5 + 32
            print(f"\n{valor}K equivale a:")
            print(f"{celsius:.2f}°C")
            print(f"{fahrenheit:.2f}°F")

        case _:
            print("Unidade inválida! Use C, F ou K.")
```

```python
# 14 Leia as quatro notas bimestrais de um aluno (0 a 10 cada). Valide todas. Calcule a média aritmética.
# Use elif para o conceito: média ≥ 9 → A, ≥ 7 → B, ≥ 5 → C, ≥ 3 → D, abaixo → E.
# Com if aninhado, exiba o status: A e B → "Aprovado", C → "Recuperação", D e E → "Reprovado".
# Exiba a média com 2 casas decimais e o conceito.

# Leitura das quatro notas
n1 = float(input("Digite a 1ª nota (0 a 10): "))
n2 = float(input("Digite a 2ª nota (0 a 10): "))
n3 = float(input("Digite a 3ª nota (0 a 10): "))
n4 = float(input("Digite a 4ª nota (0 a 10): "))

# Validação das notas
if not (0 <= n1 <= 10 and 0 <= n2 <= 10 and 0 <= n3 <= 10 and 0 <= n4 <= 10):
    print("Erro: Todas as notas devem estar entre 0 e 10!")
else:
    # Cálculo da média
    media = (n1 + n2 + n3 + n4) / 4

    # Determinar conceito (elif)
    if media >= 9:
        conceito = "A"
    elif media >= 7:
        conceito = "B"
    elif media >= 5:
        conceito = "C"
    elif media >= 3:
        conceito = "D"
    else:
        conceito = "E"

    # If aninhado: determinar status
    if conceito == "A" or conceito == "B":
        status = "Aprovado"
    elif conceito == "C":
        status = "Recuperação"
    else:  # conceito D ou E
        status = "Reprovado"

    # Exibição dos resultados
    print("\n--- BOLETIM ESCOLAR ---")
    print(f"Notas: {n1:.1f}, {n2:.1f}, {n3:.1f}, {n4:.1f}")
    print(f"Média: {media:.2f}")
    print(f"Conceito: {conceito}")
    print(f"Status: {status}")
```
