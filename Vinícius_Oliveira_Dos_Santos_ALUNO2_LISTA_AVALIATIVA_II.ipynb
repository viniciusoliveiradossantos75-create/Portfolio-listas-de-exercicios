```python
# 02 Em uma instituição de ensino, um professor deseja analisar o desempenho de uma turma sem saber previamente a quantidade de alunos.
#Desenvolva um programa que leia as notas dos alunos até que seja digitado o valor -1 (fim da entrada).
#Ao final, calcule a média da turma, informe quantos alunos foram aprovados (nota ≥ 7) e quantos foram reprovados.

notas = []
print("Digite as notas (digite -1 para sair):")

while True:
    nota = float(input("Nota: "))
    if nota == -1:
        break
    notas.append(nota)

if notas:
    media = sum(notas) / len(notas)
    aprovados = len([n for n in notas if n >= 7])
    reprovados = len(notas) - aprovados

    print(f"\nMédia: {media:.2f}")
    print(f"Aprovados: {aprovados}")
    print(f"Reprovados: {reprovados}")
else:
    print("Nenhuma nota inserida.")
```

```python
# 03 Um sistema bancário simples precisa simular operações de saque em uma conta corrente.
#O programa deve iniciar com um saldo informado pelo usuário e permitir saques sucessivos até que seja digitado 0.
#A cada tentativa, verifique se há saldo suficiente, atualize o saldo quando possível e informe ao usuário quando o valor exceder o saldo disponível.

saldo = float(input("Saldo inicial da conta: R$ "))

print(f"\nSaldo atual: R$ {saldo:.2f}")
print("Digite 0 para encerrar as operações.\n")


while True:
    valor = float(input("Valor do saque: R$ "))

    if valor == 0:
        print("\nOperações encerradas.")
        break
    elif valor > saldo:
        print(f"Saldo insuficiente! Seu saldo é R$ {saldo:.2f}")
    else:
        saldo -= valor
        print(f"Saque realizado! Novo saldo: R$ {saldo:.2f}")

print(f"\nSaldo final: R$ {saldo:.2f}")
```

```python
# 05 Um sistema de segurança controla o acesso a um ambiente restrito.
#O programa deve solicitar uma senha numérica e permitir até três tentativas.
#Caso a senha correta (2024) seja informada dentro desse limite, o acesso deve ser liberado.
#Caso contrário, após três tentativas incorretas, o sistema deve bloquear o acesso e informar essa condição ao usuário.

SENHA_CORRETA = 2024
tentativas = 0
max_tentativas = 3

print(f"Você tem {max_tentativas} tentativas")
print()

while tentativas < max_tentativas:
    tentativas += 1
    tentativas_restantes = max_tentativas - tentativas

    senha = int(input(f"Tentativa {tentativas} de {max_tentativas} - Digite a senha: "))

    if senha == SENHA_CORRETA:
        print("\n ACESSO PERMITIDO!")
        print("Bem-vindo à área restrita.")
        break
    else:
        print(f" Senha incorreta!")
        if tentativas_restantes > 0:
            print(f"Você tem mais {tentativas_restantes} tentativa(s).\n")
        else:
            print("\n ACESSO BLOQUEADO!")
            print("Número máximo de tentativas excedido.")
```

```python
#30Cálculo de Fatorial
#Um programa deve ler um número inteiro positivo e calcular seu fatorial utilizando repetição.
#Após isso, deve verificar se o resultado do fatorial é par ou ímpar e apresentar essa informação ao usuário.

numero = int(input("Digite um número inteiro positivo: "))


if numero < 0:
    print("Por favor, digite um número inteiro positivo.")
else:

    fatorial = 1


    for i in range(1, numero + 1):
        fatorial *= i

    print(f"O fatorial de {numero} é {fatorial}")

    if fatorial%2 == 0:
      print(f"fatorial resultante é par")

    else:
      print(f"fatorial resultante é impar")
```

```python
#28Caça aos Primos
#Um sistema deve analisar números inteiros informados pelo usuário até que seja digitado -1. Ao final, informe quantos números são primos e calcule a soma dos números que não são primos.

def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

prime_count = 0
non_prime_sum = 0

while True:
    num = int(input("Digite um número inteiro (-1 para sair): "))

    if num == -1:
        break

    if is_prime(num):
        prime_count += 1
    else:
        non_prime_sum += num

print(f"Quantidade de números primos: {prime_count}")
print(f"Soma dos números não primos: {non_prime_sum}")


```

```python
#26Perfil Etário
#Um sistema deve coletar a idade de diversas pessoas até que seja digitado 0.
#Ao final, deve calcular a média das idades e informar quantas pessoas são maiores de idade (acima de 18 anos).

idades = []
maiores = 0

while True:
    try:
        idade = int(input("Digite a idade (ou 0 para encerrar): "))

        if idade == 0:
            break

        idades.append(idade)

        if idade > 18:
            maiores += 1

    except ValueError:
        print("Por favor, digite um número válido.")

print("\n--- RESULTADOS ---")

if idades:
    media = sum(idades) / len(idades)
    print(f"Idades informadas: {idades}")
    print(f"Quantidade de pessoas: {len(idades)}")
    print(f"Média das idades: {media:.2f}")
    print(f"Pessoas maiores de idade (>18): {maiores}")
else:
    print("Nenhuma idade válida foi informada.")
```

```python
#06Vendas com Desconto
#Uma loja deseja automatizar o cálculo de suas vendas diárias.
#Desenvolva um programa que leia o valor de 8 vendas realizadas.
#Sempre que o valor de uma venda for superior a R$ 100, aplique automaticamente um desconto de 10% sobre esse valor.
#Ao final, informe o total arrecadado após a aplicação dos descontos.

total = 0

print("Cálculo de vendas diárias (8 vendas)")
print("-" * 40)

for i in range(1, 9):
    valor = float(input(f"Valor da {i}ª venda: R$ "))

    if valor > 100:
        valor = valor * 0.90  # Aplica desconto de 10%

    total += valor

print("-" * 40)
print(f"Total arrecadado após descontos: R$ {total:.2f}")
```

```python
#24Classificação Escolar
#Um sistema escolar precisa classificar o desempenho dos alunos com base em suas médias finais.
#O programa deve ler 8 valores e classificar cada um como aprovado, em recuperação ou reprovado.
#Ao final, apresente a quantidade de alunos em cada categoria.

print("Digite as 8 médias finais dos alunos:")

aprovados = 0
recuperacao = 0
reprovados = 0

for i in range(1, 9):
    media = float(input(f"Média do {i}º aluno: "))

    if media >= 7:
        aprovados = aprovados + 1
        print(f"  → Aprovado")
    elif media >= 5:
        recuperacao = recuperacao + 1
        print(f"  → Recuperação")
    else:
        reprovados = reprovados + 1
        print(f"  → Reprovado")

print("\n" + "="*35)
print("RELATÓRIO FINAL")
print("="*35)
print(f" Aprovados: {aprovados} aluno(s)")
print(f" Em recuperação: {recuperacao} aluno(s)")
print(f" Reprovados: {reprovados} aluno(s)")
```

```python
#15Loja de Eletrônicos
#Uma loja de eletrônicos deseja calcular o valor total das vendas realizadas em um período.
#O programa deve ler os preços dos produtos vendidos até que seja digitado 0.
#Para valores superiores a R$ 200, deve ser aplicado um desconto de 5%.
#Ao final, apresente o valor total a ser pago após os descontos.

print("Digite os preços dos produtos vendidos (digite 0 para encerrar):")

total_original = 0
total_com_desconto = 0
contador = 0

while True:
    preco = float(input("Preço do produto: R$ "))
    if preco == 0:
        break

    total_original = total_original + preco
    contador = contador + 1

    if preco > 200:
        desconto = preco * 0.05
        preco_final = preco - desconto
        print(f"  → Produto com desconto de 5%: R$ {preco_final:.2f} (original: R$ {preco:.2f})")
        total_com_desconto = total_com_desconto + preco_final
    else:
        print(f"  → Sem desconto: R$ {preco:.2f}")
        total_com_desconto = total_com_desconto + preco

print("\n" + "="*40)
print("RESUMO DAS VENDAS")
print("="*40)
print(f"Total de produtos vendidos: {contador}")
print(f"Valor total original: R$ {total_original:.2f}")
print(f"Valor total com descontos: R$ {total_com_desconto:.2f}")

if total_original > 0:
    economia = total_original - total_com_desconto
    print(f"Economia total para o cliente: R$ {economia:.2f}")
```

```python
#13Registro de Depósitos
#Um sistema financeiro deve registrar depósitos realizados em uma conta.
#O programa deve ler valores de depósitos até que seja informado o valor 0.
#Ao final, deve apresentar o total depositado, o maior valor de depósito realizado e a quantidade total de depósitos efetuados.

print("Digite os valores dos depósitos (digite 0 para encerrar):")

total_depositado = 0
maior_deposito = 0
quantidade = 0

while True:
    valor = float(input("Valor do depósito: R$ "))
    if valor == 0:
        break

    total_depositado = total_depositado + valor
    quantidade = quantidade + 1

    if valor > maior_deposito:
        maior_deposito = valor

print("\n" + "="*35)
print("EXTRATO DE DEPÓSITOS")
print("="*35)

if quantidade > 0:
    print(f"Total depositado: R$ {total_depositado:.2f}")
    print(f"Maior valor de depósito: R$ {maior_deposito:.2f}")
    print(f"Quantidade de depósitos: {quantidade}")
else:
    print("Nenhum depósito foi realizado!")
```

```python
#21Pesquisa Eleitoral
#Uma pesquisa eleitoral está sendo realizada para identificar a preferência dos eleitores.
#O programa deve ler votos até que seja digitado 0. Votos válidos são representados por códigos específicos; quaisquer outros valores devem ser considerados inválidos. Ao final, informe a quantidade de votos válidos, inválidos e o candidato vencedor.

print("Códigos dos candidatos: 1, 2, 3")
print("Digite 0 para encerrar a votação")

votos_validos = 0
votos_invalidos = 0
votos_candidato1 = 0
votos_candidato2 = 0
votos_candidato3 = 0

while True:
    voto = int(input("Digite o código do voto: "))
    if voto == 0:
        break

    if voto == 1:
        votos_candidato1 += 1
        votos_validos += 1
    elif voto == 2:
        votos_candidato2 += 1
        votos_validos += 1
    elif voto == 3:
        votos_candidato3 += 1
        votos_validos += 1
    else:
        votos_invalidos += 1
        print("  → Voto inválido!")

print("\n" + "="*35)
print("RESULTADO DA PESQUISA")
print("="*35)
print(f"Votos válidos: {votos_validos}")
print(f"Votos inválidos: {votos_invalidos}")
print(f"\nCandidato 1: {votos_candidato1} votos")
print(f"Candidato 2: {votos_candidato2} votos")
print(f"Candidato 3: {votos_candidato3} votos")

if votos_validos > 0:
    if votos_candidato1 > votos_candidato2 and votos_candidato1 > votos_candidato3:
        print(f"\n Candidato vencedor: Candidato 1")
    elif votos_candidato2 > votos_candidato1 and votos_candidato2 > votos_candidato3:
        print(f"\n Candidato vencedor: Candidato 2")
    elif votos_candidato3 > votos_candidato1 and votos_candidato3 > votos_candidato2:
        print(f"\n Candidato vencedor: Candidato 3")
    else:
        print(f"\n Empate entre os candidatos!")
else:
    print("\nNenhum voto válido foi registrado!")
```

```python
#22Estacionamento
#Um estacionamento deseja automatizar o cálculo de sua arrecadação diária.
#O programa deve ler o tempo de permanência de cada veículo até que seja digitado 0.
#O valor a ser pago deve ser calculado com base no tempo, com aplicação de desconto para permanências superiores a 5 horas.
#Ao final, apresente o total arrecadado.

print("Digite o tempo de permanência de cada veículo (em horas)")
print("Digite 0 para encerrar")

total_arrecadado = 0
veiculos_atendidos = 0

while True:
    horas = float(input("Tempo de permanência: "))
    if horas == 0:
        break

    veiculos_atendidos += 1


    valor = horas * 5


    if horas > 5:
        desconto = valor * 0.10
        valor_com_desconto = valor - desconto
        print(f"  → Valor original: R$ {valor:.2f} | Desconto: R$ {desconto:.2f} | Total: R$ {valor_com_desconto:.2f}")
        total_arrecadado += valor_com_desconto
    else:
        print(f"  → Valor a pagar: R$ {valor:.2f}")
        total_arrecadado += valor

print("\n" + "="*35)
print("RELATÓRIO DO ESTACIONAMENTO")
print("="*35)
print(f"Total de veículos atendidos: {veiculos_atendidos}")
print(f"Total arrecadado: R$ {total_arrecadado:.2f}")
```

```python
#18Consumo de Energia
#Uma concessionária de energia deseja calcular o valor total arrecadado com base no consumo de seus clientes.
#O programa deve ler o consumo de energia de 6 residências.
# Para consumos superiores a 100 kWh, deve ser aplicada uma tarifa diferenciada. Ao final, o sistema deve apresentar o total arrecadado.

print("Digite o consumo de 6 residências (em kWh)")

total_arrecadado = 0
tarifa_normal = 0.50
tarifa_diferenciada = 0.75

for i in range(1, 7):
    consumo = float(input(f"Consumo da {i}ª residência (kWh): "))

    if consumo > 100:
        valor = consumo * tarifa_diferenciada
        print(f"  → Tarifa diferenciada: R$ {valor:.2f} (R$ {tarifa_diferenciada:.2f}/kWh)")
    else:
        valor = consumo * tarifa_normal
        print(f"  → Tarifa normal: R$ {valor:.2f} (R$ {tarifa_normal:.2f}/kWh)")

    total_arrecadado += valor

print("\n" + "="*35)
print("RELATÓRIO DE ARRECADAÇÃO")
print("="*35)
print(f"Total arrecadado: R$ {total_arrecadado:.2f}")
```

```python
#16Tempo de Atendimento
#Um sistema de atendimento deve registrar o tempo gasto no atendimento de clientes em uma fila.
#O programa deve ler o tempo de atendimento de cada cliente até que seja digitado 0.
#Ao final, informe o número total de clientes atendidos e o tempo médio de atendimento.

print("Digite o tempo de atendimento de cada cliente (em minutos)")
print("Digite 0 para encerrar")

total_tempo = 0
clientes_atendidos = 0

while True:
    tempo = float(input("Tempo de atendimento: "))
    if tempo == 0:
        break

    total_tempo += tempo
    clientes_atendidos += 1

print("\n" + "="*35)
print("RELATÓRIO DE ATENDIMENTO")
print("="*35)

if clientes_atendidos > 0:
    media = total_tempo / clientes_atendidos
    print(f"Total de clientes atendidos: {clientes_atendidos}")
    print(f"Tempo total de atendimento: {total_tempo:.2f} minutos")
    print(f"Tempo médio de atendimento: {media:.2f} minutos")
else:
    print("Nenhum cliente foi atendido!")
```

```python
#10Produção Semanal
#Uma indústria deseja acompanhar sua produção semanal para avaliar o desempenho operacional.
#O programa deve ler a quantidade produzida em cada um dos 7 dias da semana e, ao final, calcular a média de produção, informar o maior valor registrado e indicar quantos dias tiveram produção inferior a 100 unidades.

print("Digite a quantidade produzida em cada um dos 7 dias:")

producao_diaria = []
dias_abaixo_100 = 0

for i in range(1, 8):
    producao = int(input(f"Produção do dia {i}: "))
    producao_diaria.append(producao)

    if producao < 100:
        dias_abaixo_100 += 1

media = sum(producao_diaria) / len(producao_diaria)
maior_producao = max(producao_diaria)

print("\n" + "="*35)
print("RELATÓRIO DE PRODUÇÃO")
print("="*35)
print(f"Média de produção: {media:.2f} unidades")
print(f"Maior produção registrada: {maior_producao} unidades")
print(f"Dias com produção inferior a 100 unidades: {dias_abaixo_100}")
```

```python
#09 Votação Eletrônica
#Um sistema de votação eletrônica simplificado será utilizado em uma eleição interna.
#Os votos devem ser informados por meio de códigos numéricos (1, 2 ou 3), sendo que o valor 0 indica o encerramento da votação.
#O programa deve contabilizar os votos de cada candidato e, ao final, informar qual candidato obteve a maior quantidade de votos.

print("Códigos dos candidatos:")
print("1 - Candidato A")
print("2 - Candidato B")
print("3 - Candidato C")
print("Digite 0 para encerrar a votação")

votos_a = 0
votos_b = 0
votos_c = 0

while True:
    voto = int(input("Digite seu voto: "))
    if voto == 0:
        break

    if voto == 1:
        votos_a += 1
        print("  → Voto confirmado para Candidato A")
    elif voto == 2:
        votos_b += 1
        print("  → Voto confirmado para Candidato B")
    elif voto == 3:
        votos_c += 1
        print("  → Voto confirmado para Candidato C")
    else:
        print("  → Voto inválido! Use os códigos 1, 2 ou 3.")

print("\n" + "="*35)
print("RESULTADO DA VOTAÇÃO")
print("="*35)
print(f"Candidato A: {votos_a} votos")
print(f"Candidato B: {votos_b} votos")
print(f"Candidato C: {votos_c} votos")
print(f"Total de votos: {votos_a + votos_b + votos_c}")


if votos_a > votos_b and votos_a > votos_c:
    print(f"\n Candidato vencedor: Candidato A com {votos_a} votos!")
elif votos_b > votos_a and votos_b > votos_c:
    print(f"\n Candidato vencedor: Candidato B com {votos_b} votos!")
elif votos_c > votos_a and votos_c > votos_b:
    print(f"\n Candidato vencedor: Candidato C com {votos_c} votos!")
elif votos_a == votos_b == votos_c == 0:
    print("\n Nenhum voto foi registrado!")
else:
    print(f"\n Empate! Nenhum candidato venceu isoladamente.")
```
